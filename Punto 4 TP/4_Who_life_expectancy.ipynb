Renombro las columnas dado que algunas tenían un espacio adelante o atrás.

```python
wle.rename(columns={
    "Life expectancy ": "Life_Expectancy",	
    "Adult Mortality": "Adult_Mortality",	
    "infant deaths": "Infant_Deaths",	
    "percentage expenditure": "Percentage_Expenditure",	
    "Hepatitis B": "Hepatitis_B",	
    "Measles ": "Measles", 
    " BMI ": "BMI", 
    "under-five deaths ": "Under_Five_Deaths",
    "Total expenditure": "Total_Expenditure",	
    " HIV/AIDS": "HIV/AIDS",
    "Diphtheria ": "Diphtheria",		
    " thinness  1-19 years": "Thinness_1-19_years",
    " thinness 5-9 years": "Thinness_5-9_years",
    "Income composition of resources": "Income_Composition_of_Resources"
    },
    inplace=True)
```

```python
round(wle.head(10), 2)
```

Sumo los valores nulos de cada columna

```python
wle.isnull().sum()
```

```python
wle.dropna(inplace=True)
```

**Matriz de correlación entre todas las variables.**

```python
plt.figure(figsize=(30,15))
sns.heatmap(wle.corr(),annot=True,cmap='Blues')
plt.show()
```

**OBSERVACIONES**:

Cuanto más oscuro es el color azul indica que la correlacion será mayor.

Caso contrario, a medida que es más claro, la correlación será menor.

Dentro de las variables que se encuentran mas correlacionadas se destacan:
*   Under_Five_Deaths con Percentage_Expenditure.
*   Percentage_Expenditure con GDP.
*   Thinness_1-19_years con Thinness_5-9_years.
*   Income_Composition_of_Resources	con Schooling.

Por otro las que menos se correlacionan son las siguientes:
*   Life_Expectancy	con Adult_Mortality.
*   Life_Expectancy con HIV/AIDS.
*   HIV/AIDS	con Year.
*   Thinness_1-19_years con BMI.
*   Thinness_5-9_years con BMI.

Otra de las variables con una fuerte correlación es School y Alcohol.

```python
sns.scatterplot(x=wle['Schooling'],y=wle['Alcohol']);
```
