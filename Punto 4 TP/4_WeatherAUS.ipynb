**Normalizo con Z**

ELimino las columnas que contienen strings para hacer la normalización.

```python
w_AUS_norm = w_AUS.drop(['Date','Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow'],axis=1)
```

```python
# Defino un diccionario para guardar la data normalizada.
w_AUS_dict = dict()

# Con un ciclo iterativo recorro el dataset iris fila por fila.
for data in w_AUS_norm.columns:
    if data == "Clase":
        w_AUS_dict[data] = w_AUS[data]
        continue
    
    # Con la librería de Numpy puedo utilizar las fórmulas de mean y std.
    mean_w_AUS = np.mean(w_AUS[data])
    std_w_AUS = np.std(w_AUS[data])
    
    # Utilizo la fórmula vista en clase: x= (x-mean(x)) / stdev(x)
    w_AUS_dict[data] = round(((w_AUS[data] - mean_w_AUS) / std_w_AUS), 2)

# Utilizo de la librería de Pandas DataFrame para imprimir los datos como una tabla de dos dimensiones.
w_AUS_normalizado = pd.DataFrame(w_AUS_dict)
print(w_AUS_normalizado)
```

**Histograma normalizado**

```python
w_AUS_normalizado.hist(figsize = (20,15));
```

**Matriz de correlación normalizada**

```python
plt.figure(figsize=(30,15))
sns.heatmap(w_AUS_normalizado.corr(),annot=True,cmap='Blues')
plt.show()
```

**OBSERVACIÓN**:

Cuanto más oscuro significa que habrá mayor correlación.

Entre las variables que mayor se correlacionan se destacan:
*   Temp3pm con MaxTemp. Esto resulta sumamente lógico dado que las temperaturas más altas suelen presentarse en los horarios cercanos al mediodía.
*   Temp9am con MinTemp. Este punto tambien es sumamente entendible debido a que por la mañana las temperaturas suelen bajar.
*   Otro dato clave que se observa es que la temperatura se correlaciona fuertemente con la evaporación.
