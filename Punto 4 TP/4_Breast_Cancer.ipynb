```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Utilizo la librería Pandas para abrir el archivo. 
breast_cancer = pd.read_csv(r"https://raw.githubusercontent.com/KevinFDI/IML-Practica-1/main/datasets/breast_cancer.csv")
```

**Normalizo con Z**

```python
# Defino un diccionario para guardar la data normalizada.
breast_cancer_dict = dict()

# Con un ciclo iterativo recorro el dataset iris fila por fila.
for data in breast_cancer.columns:
    if data == "Clase":
        breast_cancer_dict[data] = breast_cancer[data]
        continue
    
    # Con la librería de Numpy puedo utilizar las fórmulas de mean y std.
    mean_breast_cancer = np.mean(breast_cancer[data])
    std_breast_cancer = np.std(breast_cancer[data])
    
    # Utilizo la fórmula vista en clase: x= (x-mean(x)) / stdev(x)
    breast_cancer_dict[data] = round(((breast_cancer[data] - mean_breast_cancer) / std_breast_cancer), 2)

# Utilizo de la librería de Pandas DataFrame para imprimir los datos como una tabla de dos dimensiones.
breast_cancer_normalizado = pd.DataFrame(breast_cancer_dict)
print(breast_cancer_normalizado)
```

**Histograma normalizado**

```python
breast_cancer_normalizado.hist(figsize = (20,15));
```

**Matriz de correlación normalizada**

```python
plt.figure(figsize=(30,15))
sns.heatmap(breast_cancer_normalizado.corr(),annot=True,cmap='Greens')
plt.show()
```

**OBSERVACIÓN**:

A mediada  que los verdes se tornan más oscuros significa que hay una correlación positiva más fuerte.

Lo contrario ocurre a medida que los colores se tornan más claros, dado que la crrelación es menor.

Con este gráfico se observa que la mayoría de las variables se correlacionan positivamente.
